# Data Cleaning

```python
cleaned_df=df
#Removing invalid user data
cleaned_df = cleaned_df.drop(cleaned_df[cleaned_df.UserId == -1].index)
```

```python
#Removing Postage represented as Orders
cleaned_df = cleaned_df.drop(cleaned_df[cleaned_df.ItemCode == -1].index)
```

```python
cleaned_df['TotalCost'] = cleaned_df['NumberOfItemsPurchased']*cleaned_df['CostPerItem']
```

## New Customer Centric Dataframe

Since the above dataframe is unnormalized and is not customer centric, a new dataframe is created in order to make the analysis easier and accurate

Since all are in IST no worries

```python
TransactionTimes = cleaned_df.groupby('TransactionId').first()
TransactionTimes = TransactionTimes[['UserId', 'TransactionTime']].groupby('UserId')['TransactionTime'].apply(lambda x: list(x))
TransactionHours = TransactionTimes.apply(lambda x: [int(i.split()[3].replace(':','')) for i in x])
TransactionDays = TransactionTimes.apply(lambda x: [i.split()[0] for i in x])
```

```python
import datetime
MostRecentTransaction = cleaned_df.groupby('TransactionId').first()[['UserId','TransactionTime']]
MostRecentTransaction['TransactionTime'] = MostRecentTransaction['TransactionTime'].apply(lambda x: datetime.datetime.strptime(x, '%a %b %d %H:%M:%S %Z %Y'))
MostRecentTransaction = MostRecentTransaction.groupby('UserId')['TransactionTime'].apply(list)
MostRecentTransaction = MostRecentTransaction.apply(lambda x: max(x))
```

```python
MorningTransactions = TransactionHours.apply(lambda x: len([i for i in x if 60000<=int(i)<120000]))
NoonTransactions = TransactionHours.apply(lambda x: len([i for i in x if 120000<=int(i)<170000]))
EveningTransactions = TransactionHours.apply(lambda x: len([i for i in x if 170000<=int(i)< 210000]))
NightTransactions = TransactionHours.apply(lambda x: len([i for i in x if 210000<=int(i) or int(i)<60000]))
WeekendTransactions = TransactionDays.apply(lambda x: len([i for i in x if i in ['Fri', 'Sat', 'Sun']]))
WeekdayTransactions = TransactionDays.apply(lambda x: len([i for i in x if i not in ['Fri', 'Sat', 'Sun']]))
TotalTransactions = TransactionDays.apply(lambda x: len(x))
```

```python
AverageCostPerItem = cleaned_df.groupby('TransactionId').mean()
AverageCPI = AverageCostPerItem.groupby('UserId').mean()['CostPerItem']
```

```python
TotalMonetary = cleaned_df.groupby('UserId').sum()['TotalCost']
```

```python
MostRecentTransaction
```

```python
header = ['AverageCPI', 'TotalTransactions', 'TotalMonetary','MostRecentTransaction',\
          'MorningTransactions','AfternoonTransactions',\
          'EveningTransactions','NightTransactions',\
          'WeekendTransactions', 'WeekdayTransactions'
          ]
Transactions = pd.merge(AverageCPI,TotalTransactions, on='UserId')
Transactions = pd.merge(Transactions,TotalMonetary , on='UserId')
Transactions = pd.merge(Transactions,MostRecentTransaction , on='UserId')
Transactions = pd.merge(Transactions, MorningTransactions, on='UserId')
Transactions = pd.merge(pd.merge(pd.merge(Transactions,NoonTransactions,on='UserId'),EveningTransactions,on='UserId'), NightTransactions, on='UserId')
Transactions = pd.merge(pd.merge(Transactions, WeekendTransactions, on='UserId'), WeekdayTransactions, on='UserId')
Transactions.columns = header
```

```python
Transactions['CostPerTransaction'] = Transactions['TotalMonetary']/Transactions['TotalTransactions']
```

### Normalizing the values


```python
for i in ['MorningTransactions', 'AfternoonTransactions','EveningTransactions','NightTransactions','WeekendTransactions', 'WeekdayTransactions']:
    Transactions[i] = Transactions[i]/Transactions['TotalTransactions']
```

## The new cleaned dataframe contains information about transactions made by each user
| ColumnName | Description | 
| --- | --- | 
| UserId | Id of the user  as provided |
| AverageCPI | Average of all Transaction's Cost per Item of the user  |
| TotalTransactions | Total Number of transactions made  by the user |
| TotalMonetary | Total money spent on all transactions by the user |
| MostRecentTransaction | Time of Most Recent Transaction as a python datetime object |
| MorningTransactions | Ratio of Number of Transaction in the morning to Total Transactions |
| AfternoonTransactions | Ratio of Number of Transaction in the Afternoon to Total Transactions |
| EveningTransactions | Ratio of Number of Transaction in the Evening to Total Transactions |
| NightTransactions | Ratio of Number of Transaction in the Night to Total Transactions |
| WeekendTransactions | Ratio of Number of Transaction in the Weekend to Total Transactions |
| WeekdayTransactions | Ratio of Number of Transaction in the Weekday to Total Transactions |
| CostPerTransaction | Average money spent in each transaction |
